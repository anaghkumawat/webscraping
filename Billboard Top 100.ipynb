```python
response = requests.get('https://www.billboard.com/charts/hot-100')

soup = BeautifulSoup(response.text, 'lxml')

all_containers = soup.find_all('div', attrs = {'class' : 'chart-list-item'})

records = []
for a_container in all_containers:
    title = a_container.find('span', attrs = {'class' : 'chart-list-item__title-text'}).text.strip()
    artist = a_container.find('div', attrs = {'class' : 'chart-list-item__artist'}).text.strip()
    rank = a_container.find('div', attrs = {'class' : 'chart-list-item__rank'}).text.strip()
    if a_container.find('div', attrs = {'class' :'chart-list-item__lyrics'}) is not None:
        lyrics = a_container.find('div', attrs = {'class' : 'chart-list-item__lyrics'}).a['href']
    else:
        lyrics = ''
    records.append((rank, title, artist,lyrics))
```

```python
import pandas as pd
df = pd.DataFrame(records, columns=['Rank','Title','Artist','Lyrics'])
```
