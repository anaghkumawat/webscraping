```python
import requests
from bs4 import BeautifulSoup

def which_year(i):
    response = requests.get('https://www.rottentomatoes.com/top/bestofrt/?year={0}'.format(i))
    soup = BeautifulSoup(response.text, 'lxml')

    result = soup.find('table', attrs= {'class' : 'table'})

    all_containers = result.find_all('tr')

    records = []
    for a_container in all_containers[1:]:

        # Rank
        rank = a_container.find('td', attrs = {'class' : 'bold'}).text[0:-1]
        
        # Rating
        rating = a_container.find('span', attrs = {'class' : 'tMeterScore'}).text.strip()
        
        # Title
        title = a_container.find('a', attrs = {'class' : 'unstyled articleLink'}).text.strip()[0:-7]
        
        # No. of Reviews
        no_of_reviews = a_container.find('td', attrs = {'class' : 'right hidden-xs'}).text
        records.append((rank, rating, title, no_of_reviews))
    return records

top_100_movies = which_year('2018')

import pandas as pd
df = pd.DataFrame(top_100_movies, columns=['Rank', 'Rating', 'Title', 'No. of Reviews'])
```
